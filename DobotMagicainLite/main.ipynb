```python
#import package

from serial.tools import list_ports
import pointList
import pydobot
import myDobotAPI
import detectColorWithImage
import alignment
```

Dobot Magician Lite

```python
#connect dobot magician lite


available_ports = list_ports.comports()
print(f'available ports: {[x.device for x in available_ports]}')
port = available_ports[0].device

device = pydobot.Dobot(port=port, verbose=True)

device.move_to(250, 0 ,70, 0, wait=True)

```

```python
#create plallet

plallet = myDobotAPI.pallet(device)

print(plallet)

for point in plallet:
    posListTemp = {
        "x" : point[1],
        "y" : point[2],
        "z" : point[3],
        "r" : point[4]
    }
    print(posListTemp)
    device.move_to(250, 0 ,70, 0, wait=True)

    myDobotAPI.myMovJ(device, posListTemp)
    device.suck(True)
    myDobotAPI.myJump(device, pointList.greenPoint)
    device.suck(False)

```

```python
plallet1 = myDobotAPI.pallet(device)
plallet2 = myDobotAPI.pallet(device)

print(plallet1)
print(plallet2)


for i in range(0, len(plallet1)):
    pos1ListTemp = {
        "x" : plallet1[i][1],
        "y" : plallet1[i][2],
        "z" : plallet1[i][3],
        "r" : plallet1[i][4]
    }
    
    pos2ListTemp = {
        "x" : plallet2[i][1],
        "y" : plallet2[i][2],
        "z" : plallet2[i][3],
        "r" : plallet2[i][4]
    }
    

    myDobotAPI.myJump(device, pos1ListTemp)
    device.suck(True)
    
    myDobotAPI.myJump(device, pos2ListTemp)
    device.suck(False)
    
device.close()
```
